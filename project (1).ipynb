```python
# plotting a bar chart for Gender and it's count

ax = sns.countplot(x = 'Gender',data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python

sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

# Plot with a specific color or palette
color_name = "blue"

sns.barplot(x='Gender', y='Amount', data=sales_gen, color=color_name)

plt.xlabel('Gender')
plt.ylabel('Total Amount')
plt.title("GENDER VS TOTAL AMOUNT")
# Display the plot
plt.show()
```

```python
#AGE Group
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
plt.title("AGE GROUP WISE GENDER COUNT")
plt.show()
```

```python
# sales vs age group
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
# Use a single color
sns.barplot(x='Age Group', y='Amount', data=sales_age, color='red')
#  2: Use a color palette for each bar (for instance, a sequential color palette)
sns.barplot(x='Age Group', y='Amount', data=sales_age, palette='Purples')
custom_colors = ['#FF6347', '#4682B4', '#32CD32', '#FFD700']
 # Replace with colors as needed
sns.barplot(x='Age Group', y='Amount', data=sales_age, palette=custom_colors)
plt.title("SALES AMOUNT VS AGE GROUP")
plt.show()
```

```python
#state number of orders
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
# Set figure size
sns.set(rc={'figure.figsize':(15,5)})
# Define custom colors for each bar
custom_colors = sns.color_palette("Paired", 10)
# Set x positions with gaps
x_positions = np.arange(len(sales_state)) *0.5
# Plot each bar with a gap
plt.bar(x_positions, sales_state['Orders'], color=custom_colors, width=0.4)
# Set x-ticks with adjusted positions and labels
plt.xticks(x_positions, sales_state['State'], rotation=25)
plt.title("STATE WISE TOTAL ORDERS ")
plt.xlabel(" State ")
plt.ylabel("Total Orders")
plt.show()
```

```python
# total amount/sales from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
custom_colors = sns.color_palette("Paired", 10)
x_positions = np.arange(len(sales_state)) * 0.5
plt.bar(x_positions, sales_state['Amount'], color=custom_colors, width=0.4)
plt.xticks(x_positions, sales_state['State'], rotation=25)
plt.title("TOTAL AMOUNT/SALES FROM TOP 10 STATES")
plt.xlabel("States")
plt.ylabel("Amount")
plt.show()
```

```python
#marital status count total
ax = sns.countplot(data = df, x = 'Marital_Status', hue='Gender')
sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# MARITAL STATUS vs amount
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_state, x = 'Marital_Status',y= 'Amount', hue='Gender')
```

```python
# occupation figure out count
sns.set(rc={'figure.figsize': (20, 5)})
ax = sns.countplot(data=df, x='Occupation', palette="viridis")  # You can also try 'Blues', 'coolwarm', etc.

# Add bar labels
for bars in ax.containers:
    ax.bar_label(bars)

plt.show()  # Display the plot
```

```python
#comparing occupations sector

sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

# Set figure size and apply a color palette
sns.set(rc={'figure.figsize': (20, 5)})
sns.barplot(data=sales_state, x='Occupation', y='Amount', palette="viridis")

plt.show()
```

```python
#maximum and minimum product sales
sns.set(rc={'figure.figsize': (30, 8)})
ax = sns.countplot(data=df, x='Product_Category', palette='viridis')

# Add bar labels
for bars in ax.containers:
    ax.bar_label(bars)

```

```python
# amount sales of product category (max to min )
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

# Set figure size and color palette
sns.set(rc={'figure.figsize': (20, 5)})
sns.barplot(data=sales_state, x='Product_Category', y='Amount', palette='viridis')  # Example color palette

plt.show()  # Display the plot
```

```python
#top sales product id oders
# Group, sort, and get the top 10 Product_IDs by Orders
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

# Set figure size and color palette
sns.set(rc={'figure.figsize': (20, 5)})
sns.barplot(data=sales_state, x='Product_ID', y='Orders', palette='viridis')  # Example color palette

plt.show()  # Display the plot
```
